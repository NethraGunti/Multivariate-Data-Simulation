```python
%matplotlib inline
import random
import math
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
colors = ['orange','red', 'blue', 'green']
```

```python
def mean(sample):
    if len(sample) <= 1:
        return 0
    sum = 0
    for i in sample:
        sum += i
    return sum/len(sample)

def variance(sample):
    if len(sample) == 0:
        return 0
    mn = mean(sample)
    sum = 0
    for i in sample:
        sum += ((i - mn)**2)
    return (sum/(len(sample)))

def normal_data(mu, sigma, size):
    data = []
    for i in range(size):
        temp = (-2*(math.log(random.random())))**0.5
        temp2 = 2*math.pi*random.random()
        data.append(sigma*temp*math.sin(temp2) + mu)
    return data
def getTDistributionData(chiSquare, df):
    sndData = normal_data(0,1,1000)
    for i in range(1000):
        chiSquare[i] = sndData[i]/(math.sqrt(chiSquare[i]/df))
    return chiSquare
def getFDistributionData(sample1, sample2, v1, v2):
    for i in range(1000):
        sample1[i] = (sample1[i]/v1)/(sample2[i]/v2)
    return sample1
```

```python
class Plotter:
    def __init__(self, r, c, title, version, n1=-1,n2=-1):
        self.version = version
        self.rows = r
        self.cols = c
        self.title = title
        self.fig, self.axs = fig, axs = plt.subplots(r, c)
        self.fig.tight_layout()
        self.n1 = n1
        self.n2 = n2
    def hist(self,sample, color, label, title, r, c):
        _,bins,_ = self.axs[r, c].hist(sample,bins = 100, density = 1, color = color, label = label)
        self.axs[r, c].set_title(title)
        self.axs[r, c].set_ylabel('Density')
        leg = self.axs[r, c].legend()
        if self.version == 0:
            best_fit_line = stats.norm.pdf(bins, 0, 1)
        if self.version == 1:
            best_fit_line = stats.chi2.pdf(bins, self.n1)
        if self.version == 2:
            best_fit_line = stats.norm.pdf(bins,0,1)
            self.axs[r,c].plot(bins, best_fit_line, marker='.', color = 'blue')
            best_fit_line = stats.t.pdf(bins, self.n1)
        if self.version == 3:
            best_fit_line = stats.f.pdf(bins, self.n1, self.n2)
        self.axs[r, c].plot(bins, best_fit_line,marker = '.', color = 'black')
```

```python
n = 25
plotter = Plotter(int(math.sqrt(n)), int(math.sqrt(n)), 'Normal distribution for various parameter values (mu and sigma)', 0)

sqrt_n = int(math.sqrt(n))
st = np.ndarray(shape=(n,1000), dtype='float64')
for i in range(sqrt_n):
    index = i%4
    for j in range(sqrt_n):
        sample = normal_data(0,1,1000)
        plotter.hist(sample, colors[index], 'mean = {:.2f}\nvar = {:.2f}'.format(mean(sample),variance(sample)),'ST num: ' + str(i*sqrt_n +j + 1).format(0,1), i, j)
        for k in range(1000):
            st[i*sqrt_n + j][k] = sample[k]
        index+=1
        index%=4
    colors.reverse()

# we just plotted all our distributions
stSquare = np.square(st)
chiSquare = np.ndarray(shape=(1, 1000), dtype = float)
```

```python
# Degree of freedom 2, we take from ST1...ST5
count = 0
plotter = Plotter(5, 2, 'ChiSquare Distribution(df=2) for all sample distributions' , 1,n1=2)
index = 0
r=0
c=0
for i in range(5):
    index = i%4
    for j in range(i+1, 5):
        chiSquare = stSquare[i] + stSquare[j]
        plotter.hist(chiSquare, colors[index], 'mean = {:.2f}\nvar = {:.2f}'.format(mean(chiSquare),variance(chiSquare)), 'ST used: '+ str(i+1) + " " + str(j+1), r,c)
        index+=1
        index%=4
        c+=1
        if(c==2):
            r += 1
            c = 0
#The black line is the pdf itself. Mean is nearly equal to the df and variance is nearly twice of df or mean
```

```python
# Degree of freedom 4, we take from ST1...ST6
count = 0
plotter = Plotter(5, 3, 'ChiSquare Distribution (df=4) for all sample distributions' , 1, n1=4)
index = 0
r=0
c=0
for i in range(6):
    for j in range(i+1, 6):
        for k in range(j+1, 6):
            index = k%4
            for p in range(k+1, 6):
                chiSquare = stSquare[i] + stSquare[j] + stSquare[k] + stSquare[p]
                plotter.hist(chiSquare, colors[index], 'mean = {:.2f}\nvar = {:.2f}'.format(mean(chiSquare),variance(chiSquare)), 'ST used: '+ str(i+1) + " " + str(j+1) + " " + str(k+1) + " " + str(p+1), r,c)
                index+=1
                index%=4
                c+=1
                if(c==3):
                    r += 1
                    c = 0
#The black line is the pdf itself. Mean is nearly equal to the df and variance is nearly twice of df or mean
```
